```python
transactions_df = spark.read.table("processed_db.transactions")
```

###Daily Sales Summary

```python
from pyspark.sql import functions as F

daily_sales_summary_df = (
    transactions_df
        .groupBy("order_date")
        .agg(
            F.countDistinct("order_id").alias("total_orders"),
            F.sum("quantity").alias("total_quantity"),
            F.sum("amount").alias("total_revenue")
        )
        .orderBy("order_date")
)
```

###Category Sales Summary

```python


category_sales_summary_df = (
    transactions_df
        .groupBy("category")
        .agg(
            F.countDistinct("order_id").alias("orders"),
            F.sum("amount").alias("revenue")
        )
        .orderBy(F.desc("revenue"))
)
```

###Save the output to presentation container as delta table

```python
daily_sales_summary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("presentation_db.transactions_daily_sales")
```

```python
category_sales_summary_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("presentation_db.transactions_category_sales")
```
